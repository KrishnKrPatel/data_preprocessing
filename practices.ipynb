```python
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset('tips')
```

```python
    data.head()
```

```python
data[data['total_bill']>50]
```

```python
data[data['total_bill']<10]
```

```python
data[(data['total_bill'] > 20) & (data['tip'] > data['total_bill'] * 0.25)]
```
