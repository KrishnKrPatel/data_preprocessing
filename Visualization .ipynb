```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x=np.linspace(0,np.pi*4,num=200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
plt.plot(x,y,color='r',label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
z=np.cos(x)
```

```python
plt.plot(x,z,label='cos(x)')
plt.plot(x,y,label='sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x,z,label='cos(x)')
plt.plot(x,y,label='sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
#subplot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,z,'r')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,z,'r')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,z,'r')

plt.show()
```
