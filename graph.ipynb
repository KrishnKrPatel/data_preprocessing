# Graph with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data',header=None,
               sep='\s+')
```

```python
df.head()
```

```python
col_names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
```

```python
df.columns=col_names
```

```python
df.head()
```

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(df['mpg'],bins=20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=df['weight'],y=df['mpg'],alpha=0.4)
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist2d(x=df['weight'],y=df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hexbin(x=df['weight'],y=df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hexbin(x=df['weight'],y=df['mpg'],gridsize=(35,15),cmap='coolwarm')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```
