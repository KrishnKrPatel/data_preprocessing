# Graph with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data',header=None,
               sep='\s+')
```

```python
df.head()
```

```python
col_names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
```

```python
df.columns=col_names
```

```python
df.head()
```

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(df['mpg'],bins=20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=df['weight'],y=df['mpg'],alpha=0.4)
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist2d(x=df['weight'],y=df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hexbin(x=df['weight'],y=df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hexbin(x=df['weight'],y=df['mpg'],gridsize=(35,15),cmap='coolwarm')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

# Graph with Seaborn

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df['mpg'])
plt.show()
```

```python
sns.distplot(df['mpg'],kde=False,bins=20)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2,size=4,hue='cylinders')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=1,size=3,hue='cylinders',col='origin')
plt.show();
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df['weight'],df['mpg'],shade=True)
plt.show()
```

```python
sns.jointplot('weight','mpg',data=df,alpha=0.4)
plt.show()
```

```python
g=sns.PairGrid(df,diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
sns.countplot(y='cylinders',data=df)
```

```python
sns.boxplot(x='cylinders',y='mpg',data=df)
plt.show()
```

```python
sns.boxplot(x='origin',y='mpg',data=df,hue='cylinders')
plt.show()
```
